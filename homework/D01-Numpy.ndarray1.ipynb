### 1. [簡答題] 請問下列兩種將 Array 轉換成 List 的方式有何不同？

```
print('list(a): ', list(a))
print('tolist(): ', a.tolist())
```


````python
import numpy as np

a = np.random.rand(2, 3);

print('list(a): ', list(a),'當Array為多維時，只會轉換第一層為List。')
print('tolist(): ', a.tolist(),'將Array所有元素，傳換為List。')
````

### 2. 請試著在程式中印出以下三個 NdArray 的屬性並且解釋結果？（屬性：ndim、shape、size、dtype、itemsize、length、type）

```
a = np.random.randint(10, size=6) 
b = np.random.randint(10, size=(3,4)) 
c = np.random.randint(10, size=(2,3,2)) 
```

````python
# 記得先 Import 正確的套件
import numpy as np
````

````python

a = np.random.randint(10, size=6) 
print(a);
print('陣列A有多少維度',a.ndim);
print('陣列A每個維度大小',a.shape);
print('陣列A中有幾個元素',a.size);
print('陣列A中的資料型態',a.dtype);
print('陣列A中每個元素占用的空間',a.itemsize); 
print('陣列A的長度',len(a));
print('陣列A的型別',type(a));
````

````python
b = np.random.randint(10, size=(3,4)) 

print(b);
print('陣列B有多少維度 : ',b.ndim);
print('陣列B每個維度大小 : ',b.shape);
print('陣列B中有幾個元素 : ',b.size);
print('陣列B中的資料型態 : ',b.dtype);
print('陣列B中每個元素占用的空間 : ',b.itemsize);
print('陣列B的長度',len(b));
print('陣列B的型別',type(b));
````

````python
c = np.random.randint(10, size=(2,3,2)) 

print(c);
print('陣列C有多少維度 : ',c.ndim);
print('陣列C每個維度大小 : ',c.shape);
print('陣列C中有幾個元素 : ',c.size);
print('陣列C中的資料型態 : ',c.dtype);
print('陣列C中每個元素占用的空間 : ',c.itemsize);
print('陣列C的長度',len(c));
print('陣列C的型別',type(c));
````

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


````python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))
````

````python
b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))
````

````python
c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
````

````python
def tolist(iterable):
    return iterable.tolist();

print(tolist(a))
print(tolist(b))
print(tolist(c))
````
